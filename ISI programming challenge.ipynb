```python
import glob
import pandas as pd
pd.set_option('display.max_rows', 400)
directory = glob.glob('./programming_challenge/*')
class Test:
    index={}
    #In the init function we create an index for the Pfiles using a dict. We use the content of the files (words 
    #in each file) rather than the filenames as keys so as to avoid repeating words in multiple P files as well as
    #to make it easier to search through the index later when calculating score.
    def __init__(self):
        for file in directory:
            path=file.split('/')
            filename=path[2]
            if(filename[0]=='P'):
                f=open(file,'r')
                content=f.readlines()
                for line in content:
                    line=line.rstrip()
                    if line not in self.index.keys():
                        self.index[line]=[]
                    self.index[line].append(filename)
    #Use a dictionary called result initially so that checking if a file has been included in it already (line 31) 
    #is fast (constant time).
    def find_property(self,strings):
        result={}
        tuples=[]
        for string in strings:
            if string in self.index.keys():
                files=self.index[string]
                for file in files:
                    if file not in result:
                        result[file]=1
                    else:
                        result[file]+=1
        #Convert the dict to a list of tuples and sort them in descending order. Return the sorted list of tuples.
        for key in result.keys():
            tuples.append((key,result[key]))
        tuples.sort(key=lambda tup: tup[1],reverse=True)
        return tuples
obj=Test()
for file in directory:
    l=[]
    path=file.split('/')
    filename=path[2]
    if filename[0]=='s':
        print(filename)
        s=open(file,'r')
        content=s.readlines()
        for line in content:
            line=line.rstrip()
            l.append(line)
        LofT=obj.find_property(l)
        #Create a DataFrame from the list of tuples to make output cleaner and print. 
        df=pd.DataFrame(LofT,columns=['Filename','Score']) 
        print(df)
        print()
        print()
```
